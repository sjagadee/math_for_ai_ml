## Limits

The calculus og inginitesimals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) #start, finish, n points
```

if $y = x^2 + 2x + 2$

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, y)
```

* There are no straight lines on the curve. 
* If we zoom in _infinitely_ close, however, we observe curves that _approach_ lines. 
* This enables us to find a slope $m$ (tangent) anywhere on the curve, including to identify where $m = 0$: 

```python
fig, ax = plt.subplots()
ax.set_xlim([-2, 0])
ax.set_ylim([0, 2])
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.5, -0.5])
ax.set_ylim([0.75, 1.25])
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.1, -0.90])
ax.set_ylim([0.95, 1.075])
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.01, -0.99])
ax.set_ylim([0.99, 1.01])
_ = ax.plot(x, y)
```

This proves that = As integral accuracy improves as we approach an infinite sided - polygon (5 sided, 8 sided, 15 sided....) finally we reach a circle

so too in differential accuracy improves as we approach a curve infinitely closely like we have dipected above

* Limits are trivially easy to calculate for continuous function, eg:
What is the limit as $x$ approaches 5 in the expression $x^2 + 2x + 2$?

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.xlim(-5, 10)
plt.ylim(-10, 80)
plt.axvline(x=5, color="purple", linestyle="--")
plt.axhline(y=37, color="purple", linestyle="--")
_ = ax.plot(x, y)
```

Let us evaluate this equation - that is not continuous:

$\lim_{x \to 1} \frac{x^2 - 1}{x - 1}$

```python
def my_func(my_x):
    my_y = (my_x**2 -1) / (my_x - 1)
    return my_y
```

```python
my_func(2)
```

```python
# will get division by zero error = comment it
# my_func(1)
```

```python
# approaching the limit of 1 will work
my_func(0.9)
```

```python
my_func(0.999)
```

```python
my_func(1.1)
```

```python
my_func(1.001)
```

```python
y = my_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color="purple", linestyle="--")
plt.axhline(y=2, color="purple", linestyle="--")
_ = ax.plot(x, y)
```

let us calculate for another equation
$\lim_{x \to 0} \frac{\text{sin } x}{x}$

```python
def sin_func(my_x):
    my_y = np.sin(my_x) / my_x
    return my_y
```

```python
# uncomment the following line to get a division by zero error
# y = sin_func(0)
```

```python
sin_func(-0.1)
```

```python
sin_func(-0.001)
```

```python
sin_func(0.001)
```

```python
y = sin_func(x) 
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.xlim(-10, 10)
plt.ylim(-1, 2)
plt.axvline(x=0, color="purple", linestyle="--")
plt.axhline(y=1, color="purple", linestyle="--")
_ = ax.plot(x, y)
```

$ \lim_{x \to \infty} \frac{25}{x} $

```python
def inf_func(my_x):
    my_y = 25 / my_x
    return my_y
```

```python
inf_func(1e3)
```

```python
inf_func(1e6)
```

```python
y = inf_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.xlim(-10, 10)
plt.ylim(-300, 300)
_ = ax.plot(x, y)
```

```python
left_x = x[x < 0]
right_x = x[x > 0]
```

```python
left_y = inf_func(left_x)
right_y = inf_func(right_x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.xlim(-10, 10)
plt.ylim(-300, 300)
ax.plot(left_x, left_y, c="C0")
_ = ax.plot(right_x, right_y, c="C0")
```
