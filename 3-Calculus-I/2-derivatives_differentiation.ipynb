## Computing derivatives with differentiation

1. The delta method - to find the slope of the curve at any given point

Let us consider the equation $y = x^2 + 2x + 2$:


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) #start, finish, n points
```

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
_ = ax.plot(x, y)
```

Let's identify slope where, say x = 2

let us determine what $y$ is:


```python
f(2)
```

Let us call this point $P$, which is located at (2, 10):

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.scatter(2, 10) 
_ = ax.plot(x, y)
```

The _delta method_ uses the difference between 2 points to calculate slope. To illustrate this, let's define another point $Q$ where, say, $x = 5$

```python
f(5)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.scatter(2, 10) 
plt.scatter(5, 37, c='orange', zorder=3)
_ = ax.plot(x, y)
```

To find the slope $m$ between points $P$ and $Q$:

$m = \frac{\text{change in }y}{\text{change in }x} = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{37-10}{5-2} = \frac{27}{3} = 9$

```python
m = (37-10)/(5-2)
m
```

- we represent $b$ that is the point where the line crosses through the y-axis
- To plot the line that passes through $P$ and $Q$, we can rearrange the equation of the line $y = mx +b$ to solve for b:


$b = y - mx$


```python
b = 37 - m * 5
b
```

```python
line_y = m * x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.scatter(2, 10) 
plt.scatter(5, 37, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x, y)
```

The closer $Q$ becomes to $P$, the closer the slope $m$ comes to being the true tangent to the point $P$. Let us demonstrate that with another point $Q$:
- where $x = 2.1$

Previously, our $\Delta x$ between $P$ and $Q$ was equal to 3, now it is much smaller:

$\Delta x = x_2 - x_1 = 2.1 - 2 = 0.1$

```python
f(2.1)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.scatter(2, 10) 
plt.scatter(2.1, 10.61, c='orange', zorder=3)
_ = ax.plot(x, y)
```

```python
m = (10.61-10)/(2.1-2)
m
```

```python
# b = y - m * x - the y intercept of the line b
b = 10.61 - m * 2.1
b
```

```python
line_y = m * x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.scatter(2, 10) 
plt.scatter(2.1, 10.61, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x, y)
```

The closer $Q$ becomes to $P$ (i.e., $\Delta x$ approaches 0), the clearer it becomes that the slope $m$ at point $P$ = (2, 10) is equal to 6.

Let's make $\Delta x$ extremely small, 0.000001, to illustrate this:

```python
delta_x = 0.000001
delta_x
```

```python
x1 = 2
y1 = 10
```

Rearranging $\Delta x = x_2 - x1$ we can calculate $x_2$ for our point $Q$, which is now extremely close to $P$:

$x_2 = x_1 + \Delta x$


```python
x2 = x1 + delta_x
x2
```

\$y_2$ for our point $Q$ can be obtained with the usual function $f(x)$: 

$y_2 = f(x_2)$

```python
y2 = f(x2)
y2
```

To find the slope $m$, we continue to use 

$m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}$

```python
m = (y2-y1)/(x2-x1)
m
```

Boom! Using the delta method, we've shown that at point $P$, the slope of the curve is 6.

Point $Q$ is at (-0.999999, 1.000000000001), extremely close to $P$.

Therefore, as $x_2$ becomes infinitely close to $x_1$, it becomes clear that the slope $m$ at $x_1 = -1$ is equal to zero. Let's plot it out: 

```python
b = y2 - m * x2
b
```

```python
line_y = m * x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")
plt.scatter(x1,y1) 
plt.scatter(x2, y2, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x, y)
```

As $Q$ becomes infinitely close to $P$:
* $x_2$ - $x_1$ approaches 0
* In other words, $\Delta x$ approaches 0
* This can be denoted as $\Delta x \to 0$

Using the delta method, we've derived the definition of differentiation from first principles. The derivative of $y$ (denoted $dy$) with respect to $x$ (denoted $dx$) can be represented as: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$

Expanding $\Delta y$ out to $y_2 - y_1$: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{y_2 - y_1}{\Delta x}$$

Finally, replacing $y_1$ with $f(x)$ and replacing $y_2$ with $f(x + \Delta x)$, we obtain a common representation of differentiation:
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

Let's observe the differentiation equation in action:

```python
def diff_demo(my_f, my_x, my_delta_x):
    return (my_f(my_x + my_delta_x) - my_f(my_x)) / my_delta_x
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(f, -1, delta))
```
