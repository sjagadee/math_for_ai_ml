Example 1: Bank robber and police example
1. Robber going at 150 KM/H in a car - with 5 mins head start
2. Police going at 180 KM?H in a car
3. How long will the police take to catch up with the robber - find time
4. What is the distance they would have travelled when the police catches up with the robber

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) ## start, end, number of points 
```

Distance travelled by robber (150 Km / h) = 2.5 km/m 

d = 2.5 * t

```python
d_r = 2.5 * t
```

Distance travelled by police at (180 Km / h) = 3 km/m - after 5 mintues

d =  3 * (t - 5)

```python
d_p = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title("Bank robber caught")
plt.xlabel("Time (min)")
plt.ylabel("Distance (km)")

ax.set_xlim(0, 40)
ax.set_ylim(0, 100)

ax.plot(t, d_r, label="Robber", color="red")
ax.plot(t, d_p, label="Police", color="blue")

plt.axvline(x=30, color="purple", linestyle="--")
_ = plt.axhline(y=75, color="purple", linestyle="--")

ax.legend()
plt.show()
```

## Conclusion 

1. Time elapsed - 30 minutes
2. Distance travelled - 75 KM
