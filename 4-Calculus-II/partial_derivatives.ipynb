### Partial Derivatives of Multivariate functions

define a function $f(x, y)$ for $z = x^2 - y^2$:

```python
def f(my_x, my_y):
    return my_x**2 - my_y**2
```

Plot $z$ with respect to $x$ by varying $x$

```python
xs = np.linspace(-3, 3, 1000)
```

...while holding y as constant (eg: at $y$ = 0)

```python
zs_wrt_x = f(xs, 0)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

plt.xlabel("x")
plt.ylabel("z", rotation=0)
_ = ax.plot(xs, zs_wrt_x)
```

To determine the slope of $z$ w.r.t. $x$ at a given point along the curve, we can use the partial derivative from the slides: $$ \frac{\partial z}{\partial x} = 2x$$

```python
## y is not relevant for this perticular derivativel it often would be
def delz_delx(my_x, my_y):
    return 2 * my_x
```

```python
x_samples = [-2, -1, 0, 1, 2]
colors = ["red", "orange", "blue", "teal", "purple"]
```

```python
def point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):

    my_z = my_f(my_x, my_y) # z = f(x, y)
    plt.scatter(my_x, my_z, color=col, zorder=3)

    tangent_m = fprime(my_x, my_y) # slope of partial derivative of f(x, y) w.r.t. x
    tangent_b = my_z - tangent_m * my_x # Line is z = mx + b, so b = z - mx
    tangent_line = tangent_m * my_xs + tangent_b

    plt.plot(my_xs, tangent_line, color=col, linestyle="--", linewidth=0.7, zorder=3)

```

```python
fig, ax = plt.subplots()

plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

for i, x in enumerate(x_samples):
    point_and_tangent_wrt_x(xs, x, 0, f, delz_delx, colors[i])

plt.ylim(-3, 9)
plt.xlabel("x")
plt.ylabel("z", rotation=0)
_ = ax.plot(xs, zs_wrt_x)


```

Thereby visually demonstrating $\frac{\partial z}{\partial x} = 2x$. 

That is, the slope of $z$ along the $x$ axis is *twice* the $x$ value.

Now let's plot $z$ w.r.t. $y$ by varying y..

```python
ys = np.linspace(-3, 3, 1000)

# while holding x constant at 0
zs_wrt_y = f(0, ys)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

plt.xlabel("y")
plt.ylabel("z", rotation=0)
_ = ax.plot(ys, zs_wrt_y)
```

The partial derivative of $z$ w.r.t $y$ happens to be independent of $x$ (just as we observed $x$ is independent of $y$ above), so while $z$ varies as a function of both $x$ and $y$, the slope of $z$ w.r.t $y$ is the same no matter what $x$ is: $$ \frac{\partial z}{\partial y} = -2y $$

So for example, holding $x$ constant at 2 instead of 0 increases $z$, but has no impact whatsoever on the slope of $z$ w.r.t. $y$: 

```python
zs_wrt_y = f(2, ys)

fig, ax = plt.subplots()
plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

plt.xlabel("y")
plt.ylabel("z", rotation=0)
_ = ax.plot(ys, zs_wrt_y)
```

```python
def delz_dely(my_x, my_y):
    return -2 * my_y
```

```python
y_samples = [-2, -1, 0, 1, 2]
```

```python
def point_and_tangent_wrt_y(my_ys, my_x, my_y, my_f, fprime, col):

    my_z = my_f(my_x, my_y) # z = f(x, y)
    plt.scatter(my_y, my_z, color=col, zorder=3)

    tangent_m = fprime(my_x, my_y) # slope of partial derivative of f(x, y) w.r.t. y
    tangent_b = my_z - tangent_m * my_y # replace x with y
    tangent_line = tangent_m * my_ys + tangent_b

    plt.plot(my_ys, tangent_line, color=col, linestyle="--", linewidth=0.7, zorder=3)
```

```python
fig, ax = plt.subplots()

plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

for i, y in enumerate(y_samples):
    point_and_tangent_wrt_y(ys, 2, y, f, delz_dely, colors[i])

plt.ylim(-5, 5)
plt.xlabel("y")
plt.ylabel("z", rotation=0)
_ = ax.plot(ys, zs_wrt_y)
```

Thereby visually demonstrating $\frac{\partial z}{\partial y} = -2y$. 

That is, the slope of $z$ along the $y$ axis is *twice* the $y$ value *and inverted*, resulting in the parabolic curve opening downward.
